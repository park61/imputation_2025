```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the provided data
df = pd.DataFrame({
    'n': [100, 100, 100, 500, 500, 500, 1000, 1000, 1000],
    'algorithm': [
        'Missforest', 'softImpute', 'monoImpute',
        'Missforest', 'softImpute', 'monoImpute',
        'Missforest', 'softImpute', 'monoImpute'
    ],
    'acc': [
        0.4523, 0.4349, 0.5263,
        0.4254, 0.4285, 0.4940,
        0.4193, 0.4194, 0.4836
    ]
})

# Pivot the data to get 'n' as the index and algorithms as columns
df_pivot = df.pivot(index='n', columns='algorithm', values='acc')

# Reindex the columns to maintain the desired order
df_pivot = df_pivot[['Missforest', 'softImpute', 'monoImpute']]

# Plot a bar graph
ax = df_pivot.plot(kind='bar', figsize=(10, 7), legend=True)

# Set labels and title
ax.set_xlabel('Number of Movies (n)')
ax.set_ylabel('Acc')
ax.set_title('Acc Comparison of Algorithms')
ax.legend(title='Algorithm')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with the provided data
df = pd.DataFrame({
    'n': [100, 100, 100, 500, 500, 500, 1000, 1000, 1000],
    'algorithm': [
        'Missforest', 'softImpute', 'monoImpute',
        'Missforest', 'softImpute', 'monoImpute',
        'Missforest', 'softImpute', 'monoImpute'
    ],
    'time': [
       316.1740712,	3.82777602,	33.3692116,
       562.5410702,	23.15486476,	102.0672156,
       678.6128116,	42.98865964,	154.1231736
    ]
})

# Pivot the data to get 'n' as the index and algorithms as columns
df_pivot = df.pivot(index='n', columns='algorithm', values='time')

# Reindex the columns to maintain the desired order
df_pivot = df_pivot[['Missforest', 'softImpute', 'monoImpute']]

# Plot a bar graph
ax = df_pivot.plot(kind='bar', figsize=(10, 7), legend=True)

# Set labels and title
ax.set_xlabel('Number of Movies (n)')
ax.set_ylabel('Time (sec.)')
ax.set_title('Average Solving-time Comparison of Algorithms')
ax.legend(title='Algorithm')

# Show the plot
plt.tight_layout()
plt.show()
```
